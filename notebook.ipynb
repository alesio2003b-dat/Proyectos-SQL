Si tenemos en cuenta el calor necesario para la fabricación y el transporte de productos, _las emisiones de gases de efecto invernadero atribuibles a los productos —desde alimentos hasta zapatillas o electrodomésticos— representan más del 75 % de las emisiones globales._ (`Fuente: The Carbon Catalogue https://www.nature.com/articles/s41597-022-01178-9`)

Nuestros datos, disponibles públicamente en nature.com, contienen las huellas de carbono de productos (PCF, por sus siglas en inglés) de varias empresas. Las PCF son las emisiones de gases de efecto invernadero atribuibles a un producto concreto, medidas en CO<sub>2</sub> equivalente (dióxido de carbono equivalente).
<!--https://www.nature.com/articles/s41597-022-01178-9-->

Estos datos están almacenados en una base de datos PostgreSQL con una única tabla, `product_emissions`, que analiza las PCF por producto y por la etapa del proceso de producción en la que se generaron dichas emisiones. Aquí tienes una vista rápida de lo que contiene `product_emissions` en cada columna:

### `product_emissions`

| field                              | data type |
|------------------------------------|-----------|
| `id`                                 | `VARCHAR`   |
| `year`                               | `INT`       |
| `product_name`                       | `VARCHAR`   |
| `company`                            | `VARCHAR`   |
| `country`                            | `VARCHAR`   |
| `industry_group`                     | `VARCHAR`   |
| `weight_kg`                          | `NUMERIC`   |
| `carbon_footprint_pcf`               | `NUMERIC`   |
| `upstream_percent_total_pcf`         | `VARCHAR`   |
| `operations_percent_total_pcf`       | `VARCHAR`   |
| `downstream_percent_total_pcf`       | `VARCHAR`   |

¡Usarás estos datos para analizar la huella de carbono de cada sector de la industria en el conjunto de datos!

```python
-- Update your query here
SELECT *
FROM product_emissions
LIMIT 5;
```

```python
select industry_group, 
	count(Distinct Company) as num_companies,
	round(sum(carbon_footprint_pcf),1) as total_industry_footprint
from product_emissions
	where year=(select max(year) from product_emissions)
group by (industry_group)
order by total_industry_footprint desc
```
